## **Cleaning Data**

### **Menangani Masalah Data type dari Dataset Orders**

```python
# Untuk mengubah tipe data yang tidak sesuai
datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])

# Untuk mengecek apakah tipe data sudah sesuai atau belum
orders_df.info()
```

### **Menangani Masalah Data Type dari Dataset Order Items**

```python
# Untuk mengubah tipe data yang tidak sesuai
datetime_columns = ["shipping_limit_date"]
for column in datetime_columns:
  order_items_df[column] = pd.to_datetime(order_items_df[column])

# Untuk mengecek apakah tipe data sudah sesuai atau belum
order_items_df.info()
```

```python
# Menghapus baris yang memiliki missing value pada order_approved_at hanya jika statusnya 'canceled'
orders_df = orders_df[~(orders_df['order_approved_at'].isna() & (orders_df['order_status'] == 'canceled'))]
# Cek apakah masih ada missing value di order_approved_at
orders_df[orders_df['order_approved_at'].isna()]
```

# **Visualization & Explanatory Analysis**

## **Pertanyaan 1: Kota mana yang memiliki jumlah transaksi (order) terbanyak?**

```python
# Menghitung jumlah transaksi per kota
city_orders = customers_df.merge(orders_df, on='customer_id').groupby('customer_city')['order_id'].count()

# Mengambil 5 kota dengan transaksi terbanyak dan tersedikit
top_5_cities = city_orders.nlargest(5)
bottom_5_cities = city_orders.nsmallest(5)

# Membuat dua canvas untuk dua bar chart
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot untuk 5 kota dengan transaksi terbanyak
sns.barplot(x=top_5_cities.values, y=top_5_cities.index, palette="Blues", ax=axes[0])
axes[0].set_title("5 Kota dengan Transaksi Terbanyak")
axes[0].set_xlabel("Jumlah Transaksi")
axes[0].set_ylabel("Kota")

# Plot untuk 5 kota dengan transaksi tersedikit
sns.barplot(x=bottom_5_cities.values, y=bottom_5_cities.index, palette="Reds", ax=axes[1])
axes[1].set_title("5 Kota dengan Transaksi Tersedikit")
axes[1].set_xlabel("Jumlah Transaksi")
axes[1].set_ylabel("Kota")

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```

## **Pertanyaan 2: Kategori produk apa yang paling banyak dibeli oleh pelanggan?**

```python
# Menghitung jumlah produk yang dibeli per kategori
product_orders = order_items_df.merge(products_df, on='product_id').groupby('product_category_name')['order_id'].count()

# Mengambil 4 kategori produk yang paling banyak dibeli
top_4_categories = product_orders.nlargest(4)

# Mengambil 4 kategori produk yang paling sedikit dibeli
bottom_4_categories = product_orders.nsmallest(4)

# Warna untuk pie chart
colors_top = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
colors_bottom = ['#c2c2f0', '#ffb3e6', '#ff6666', '#99e6e6']

# Membuat figure dengan dua subplot untuk pie chart
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart untuk 4 Kategori Produk Paling Banyak Dibeli
axes[0].pie(top_4_categories.values, labels=top_4_categories.index, autopct='%1.1f%%', colors=colors_top, startangle=140, wedgeprops={'edgecolor': 'black'})
axes[0].set_title("4 Kategori Produk Paling Banyak Dibeli")

# Pie chart untuk 4 Kategori Produk Paling Sedikit Dibeli
axes[1].pie(bottom_4_categories.values, labels=bottom_4_categories.index, autopct='%1.1f%%', colors=colors_bottom, startangle=140, wedgeprops={'edgecolor': 'black'})
axes[1].set_title("4 Kategori Produk Paling Sedikit Dibeli")

# Menyesuaikan layout agar tidak bertabrakan
plt.tight_layout()
plt.show()
```

## **Pertanyaan 3: Kategori produk apa yang paling besar menghasilkan omset?**

```python
# Menghitung total omzet per kategori produk
omzet_per_kategori = products_order_items_df.groupby("product_category_name", as_index=False)["price"].sum()

# Mengurutkan omzet dari yang terbesar
omzet_per_kategori = omzet_per_kategori.sort_values(by="price", ascending=False).head(10)

# Membuat figure
plt.figure(figsize=(12, 6))
sns.barplot(y=omzet_per_kategori["product_category_name"],
            x=omzet_per_kategori["price"],
            palette="viridis")

# Menambahkan label
plt.xlabel("Total Omzet", fontsize=12)
plt.ylabel("Kategori Produk", fontsize=12)
plt.title("Kategori Produk dengan Omzet Terbesar", fontsize=14)

# Menampilkan nilai di setiap bar
for i, value in enumerate(omzet_per_kategori["price"]):
    plt.text(value, i, f"{value:,.0f}", ha="left", va="center", fontsize=10)

# Menampilkan plot
plt.show()
```

## **Pertanyaan 4: Penjual mana yang memiliki jumlah transaksi (order) terbanyak?**

```python
# Menghitung jumlah order per penjual
seller_order_counts = order_items_df.groupby('seller_id')['order_id'].nunique().sort_values(ascending=False)

# Mengambil 10 seller dengan jumlah order terbanyak
top_sellers = seller_order_counts.head(10)

# Menghitung persentase kumulatif
cumulative_percentage = np.cumsum(top_sellers) / top_sellers.sum() * 100

# Membuat figure
fig, ax1 = plt.subplots(figsize=(12, 6))

# Membuat bar chart (jumlah order)
sns.barplot(x=top_sellers.index, y=top_sellers.values, ax=ax1, color="skyblue")
ax1.set_ylabel("Jumlah Order", fontsize=12)
ax1.set_xlabel("Seller ID", fontsize=12)
ax1.set_title("Seller ID dengan Jumlah Order Terbanyak", fontsize=14)

# Memutar label sumbu X sebesar 45 derajat
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')

# Membuat line chart (persentase kumulatif)
ax2 = ax1.twinx()
ax2.plot(top_sellers.index, cumulative_percentage, color="red", marker="o", linestyle="-", linewidth=2)
ax2.set_ylabel("Persentase Kumulatif (%)", fontsize=12)
ax2.set_ylim([0, 110])

# Menambahkan garis horizontal untuk 80% (Prinsip Pareto)
ax2.axhline(y=80, color="gray", linestyle="--", linewidth=1)
ax2.text(0.5, 82, "80% Threshold", color="gray", fontsize=12)

# Menampilkan nilai di atas bar
for i, val in enumerate(top_sellers.values):
    ax1.text(i, val + 5, str(val), ha='center', fontsize=10)

# Menampilkan persentase kumulatif di atas titik
for i, val in enumerate(cumulative_percentage):
    ax2.text(i, val + 2, f"{val:.1f}%", ha='center', fontsize=10, color="red")

# Menampilkan plot
plt.show()
```

## **Insight**

* Kota dengan jumlah transaksi terbanyak menunjukkan area dengan tingkat belanja online yang tinggi.
* Kota dengan jumlah transaksi terendah bisa menjadi target pemasaran yang lebih baik.
* Kategori produk dengan jumlah transaksi tertinggi mencerminkan kebutuhan utama pelanggan di marketplace ini.
* Kategori dengan transaksi rendah bisa menjadi potensi pasar baru yang perlu dieksplorasi.
* Kategori dengan omzet tertinggi bukan selalu kategori dengan jumlah transaksi terbanyak.
* Beberapa kategori mungkin memiliki harga rata-rata produk yang lebih tinggi, sehingga menghasilkan omzet besar meskipun jumlah transaksi lebih sedikit.
* 20% penjual teratas berkontribusi terhadap hampir 80% total order (Prinsip Pareto).
* Ada perbedaan antara kategori paling banyak dibeli dan kategori dengan omzet terbesar → Harga produk sangat memengaruhi total omzet.
* Wilayah dengan transaksi rendah bisa menjadi peluang pemasaran baru untuk meningkatkan penetrasi pasar.

# **Analysis Lanjutan**

RFM Analysis, bertujuan mengelompokkan pelanggan berdasarkan perilaku pembelian mereka dengan memperhatikan tiga faktor utama:
*   Recency: Menghitung jumlah hari sejak terakhir kali pelanggan melakukan pembelian.
*   Frequency: Menghitung jumlah total transaksi yang dilakukan oleh pelanggan dalam periode tertentu.
*   Monetary: Menghitung total pengeluaran pelanggan dalam periode tersebut.

## **Recency**
jumlah hari sejak terakhir kali pelanggan melakukan pembelian.

```python
# Pastikan orders_df memiliki kolom 'customer_id' dan 'order_purchase_timestamp'
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])

# Menentukan tanggal acuan (misalnya, 1 hari setelah tanggal terakhir transaksi)
reference_date = orders_df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

# Menghitung Recency: selisih antara reference_date dengan tanggal transaksi terakhir setiap pelanggan
recency_df = orders_df.groupby('customer_id').agg(
    last_purchase=('order_purchase_timestamp', 'max')
).reset_index()

recency_df['Recency'] = (reference_date - recency_df['last_purchase']).dt.days

# Menampilkan hasil
recency_df = recency_df[['customer_id', 'Recency']]
print(recency_df.head())
```

```python
# Membuat figure
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram untuk melihat distribusi Recency
sns.histplot(recency_df['Recency'], bins=30, kde=True, ax=axes[0], color="skyblue")
axes[0].set_title("Distribusi Recency")
axes[0].set_xlabel("Recency (Hari)")
axes[0].set_ylabel("Jumlah Pelanggan")

# Boxplot untuk melihat outlier
sns.boxplot(x=recency_df['Recency'], ax=axes[1], color="lightcoral")
axes[1].set_title("Boxplot Recency")
axes[1].set_xlabel("Recency (Hari)")

# Menampilkan plot
plt.tight_layout()
plt.show()
```

## **Frequency**
Seberapa sering pelanggan melakukan transaksi

```python
# Menghitung Frequency (Jumlah transaksi per pelanggan)
frequency_df = orders_df.groupby("customer_id")["order_id"].nunique().reset_index()
frequency_df.columns = ["customer_id", "Frequency"]

# Melihat 5 data teratas
frequency_df.head()
```
